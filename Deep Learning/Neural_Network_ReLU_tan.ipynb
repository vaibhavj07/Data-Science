```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-2, 2, 20)
```

```python
def relu(values):
    return np.maximum(values, 0)
```

```python
relu_y = relu(x)
```

```python
plt.plot(x, relu_y)
plt.show()
```
