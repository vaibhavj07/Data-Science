```python
def centers(data):
    mean, median, modes = data.mean(), data.median(), data.mode()
    mn = plt.axvline(mean, color='red')
    md = plt.axvline(median, color='green')
    for mode in modes:
        mo = plt.axvline(mode, color='yellow')
    plt.legend((mn,md,mo), 'Mean Median Mode'.split())
```

```python
sns.distplot(iris.sepal_width)
centers(iris.sepal_width)
```

```python
sns.barplot(vals, prob)
```

```python
def discrete_prbs(data):
    vals, freqs = np.unique(data, return_counts = True)
    probs = freqs/len(data)
    return vals, probs
```

```python
# Effect Of n on Bionomial Distribution
def plot_bionomial(num_trials):
    fig, ax = plt.subplots()
    
    for n in num_trials:
        data = np.random.binomial(n, 0.5, size=10000)
        vals, probs = discrete_prbs(data)
        sns.scatterplot(vals, probs, ax=ax)
        
    plt.legend([f'n={n}' for n in num_trials])
    plt.show()
```

```python
num_trials = [10,20,50]
plot_bionomial(num_trials)
```

```python
# Poisson Distribution
data = np.random.poisson(lam = 5, size = 10000)
vals, probs = discrete_prbs(data)
sns.scatterplot(vals, probs);
```

```python
# for a lambda of 5, the distribution is positively skewed. 
```

```python
def plot_poisson(lams):
    fig, ax = plt.subplots()
    
    for lam in lams:
        data =  np.random.poisson(lam, size = 10000)
        vals, probs = discrete_prbs(data)
        sns.scatterplot(vals, probs, ax = ax)
        
    plt.legend([f'$\lambda$={lam}' for lam in lams])
    plt.show()
```

```python
lams = [1, 5, 20]
plot_poisson(lams)
```
