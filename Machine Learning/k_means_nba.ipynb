```python
# centroid based clustering. Works best when clusters resemble circles that have centers. Arithmatic mean of centers. 

#K means 
```

```python
# step 1 --> assign points to clusters
# step 2 --> update new centroids of the clu
```

```python
num_clusters = 5

random_initial_points = np.random.choice(point_gaurds.index, size=num_clusters)
```

```python
centroids = point_gaurds.loc[random_initial_points]
```

```python
centroids
```

```python
plt.figure(figsize = [10, 8])
plt.scatter(point_gaurds['ppg'], point_gaurds['atr'])
plt.scatter(centroids['ppg'], centroids['atr'])
plt.title("centroids")
plt.xlabel("points per game")
plt.ylabel("assist turnover ratio")
plt.show()
```

```python
# Key = cluster_id
# value = centroids coordinates expressed as list [ppg, atr]
```

```python
def centroids_to_dict(centroids):
    dictc = {}
    counter = 0
    
    for index, row in centroids.iterrows():
        coord = [row['ppg'], row['atr']]
        dictc[counter] = coord
        counter = counter + 1
        
    return dictc
```

```python
centroids_dict = centroids_to_dict(centroids)
```

```python
centroids_dict
```

```python
def calculate_distance(centroid, player_values):
    root_distance = 0
    
    for x in range(len(centroid)):
        difference = centroid[x] - player_values[x] 
        squared_difference = difference**2
        root_distance += squared_difference
        
    euclid_distance = math.sqrt(root_distance)
    return euclid_distance
```

```python
def assign_to_cluster(row):
    lowest_distance = None
    closest_cluster = None
    
    for cluster_id , centroid in centroids_dict.items():
        df_row = [ row['ppg'], row['atr']]
        euclidean_distance = calculate_distance(centroid, df_row)
        
        if lowest_distance is None:
            lowest_distance = euclidean_distance
            closest_cluster = cluster_id
        elif euclidean_distance < lowest_distance:
            lowest_distance = euclidean_distance
            closest_cluster = cluster_id
    return closest_cluster
```

```python
point_gaurds['cluster'] = point_gaurds.apply(lambda row: assign_to_cluster(row), axis=1)
```

```python
def visualize_clusters(df, num_clusters):
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    
    for n in range(num_clusters):
        clustered_df = df[df['cluster']==n]
        plt.scatter(clustered_df['ppg'], clustered_df['atr'], c=colors[n-1])
        plt.xlabel('points per game')
        plt.ylabel("assist turnover ratio")
    plt.show()
        
```

```python
visualize_clusters(point_gaurds, 5)
```

```python
def recalculate_centroids(df):
    new_centroids_dict = {}
    
    for cluster_id in range(0, num_clusters):
        values_in_cluster = df[df['cluster'] == cluster_id] 
        
        new_centroid = [np.average(values_in_cluster['ppg']), np.average(values_in_cluster['atr'])]
        new_centroids_dict[cluster_id] = new_centroid
        
    return new_centroids_dict
```

```python
centroids_dict = recalculate_centroids(point_gaurds)
```

```python
point_gaurds['cluster'] = point_gaurds.apply(lambda row: assign_to_cluster(row), axis=1)

visualize_clusters(point_gaurds, 5)
```

```python
centroids_dict = recalculate_centroids(point_gaurds)

point_gaurds['cluster'] = point_gaurds.apply(lambda row: assign_to_cluster(row), axis=1)

visualize_clusters(point_gaurds, 5)
```

```python
centroids_dict = recalculate_centroids(point_gaurds)

point_gaurds['cluster'] = point_gaurds.apply(lambda row: assign_to_cluster(row), axis=1)

visualize_clusters(point_gaurds, 5)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(point_gaurds[['ppg', 'atr']])

point_gaurds['cluster'] = kmeans.labels_

visualize_clusters(point_gaurds, num_clusters)
```
