```python
dfa_random = dfa.sample(frac=1, random_state = 1)
```

```python
split = round(len(dfa_random)*0.8)

split
```

```python
training_set = dfa_random[:split].reset_index(drop=True)
testing_set = dfa_random[split:].reset_index(drop=True)
```

```python
training_set['SMS'] = training_set['SMS'].str.replace('\W', ' ')
training_set['SMS'] = training_set['SMS'].str.lower()
```

```python
training_set['SMS']
```

```python
training_set['SMS'] = training_set['SMS'].str.split()

vocabulary = set()

for sms in training_set['SMS']:
    for word in sms:
        vocabulary.add(word)
        
        
len(vocabulary)
```

```python
word_counts_per_sms = {unique_word:[0]*len(training_set['SMS']) for unique_word in vocabulary}

for index, sms in enumerate(training_set['SMS']):
    for word in sms:
        word_counts_per_sms[word][index]+=1
        
```

```python
word_counts = pd.DataFrame(word_counts_per_sms)
```

```python
word_counts
```

```python
training_set_clean = pd.concat([training_set, word_counts], axis=1)
```

```python
training_set_clean
```

```python
spam_messages = training_set_clean[training_set_clean['label'] == 'spam']
ham_messages = training_set_clean[training_set_clean['label'] == 'ham']
```

```python
p_spam = len(spam_messages)/len(training_set_clean)
p_ham = len(ham_messages)/len(training_set_clean)
```

```python
# number of spam
n_words_per_spam_message = spam_messages['SMS'].apply(len)
n_spam = n_words_per_spam_message.sum()

# number of ham
n_words_per_ham_message = ham_messages['SMS'].apply(len)
n_ham = n_words_per_ham_message.sum()

# number of vocabulary
n_vocabulary = len(vocabulary)

# additive smoothing 
alpha = 1
```

```python
#p(w1|spam)
#p(w2|spam)
#..
#p(wn|spam)



#p(w1|ham)
#p(w2|ham)
#..
#p(wn|ham)
```

```python
parameters_spam = {unique_word:0 for unique_word in vocabulary}
parameters_ham = {unique_word:0 for unique_word in vocabulary}
```

```python
# calculation parameters
```

```python
for word in vocabulary:
    n_word_given_spam = spam_messages[word].sum()
    p_word_given_spam = (n_word_given_spam + alpha)/(n_spam + alpha*n_vocabulary)
    parameters_spam[word] = p_word_given_spam
    
    n_word_given_ham = ham_messages[word].sum()
    p_word_given_ham = (n_word_given_ham + alpha)/(n_ham + alpha*n_vocabulary)
    parameters_ham[word] = p_word_given_ham
    
```

```python
import re

def spam_ham(msg):
        msg = re.sub('\W', ' ', msg)
        msg = msg.lower().split()
        
        p_spam_given_msg = p_spam
        p_ham_given_msg = p_ham
        
        for word in msg:
            if word in parameters_spam:
                p_spam_given_msg *= parameters_spam[word]
            if word in parameters_ham:
                p_ham_given_msg *= parameters_ham[word]
        if p_ham_given_msg >= p_spam_given_msg:
            label = 'ham'
        elif p_ham_given_msg < p_spam_given_msg:
            label = 'spam'
        print(f"the message is {label}")
        
        return label
```

```python
spam_ham("you get a discount on amazon")
```

```python
spam_ham("Free entry in 2 a wkly comp to win FA Cup fina")
```

```python
testing_set['prediction'] = testing_set['SMS'].apply(spam_ham)
```

```python
testing_set
```

```python
correct = 0
total = len(testing_set)


for row in testing_set.iterrows():
    row = row[1]
    if row['label'] == row['prediction']:
        correct+=1
        
print(f"correct--> {correct}")
print(f"incorrect--> {total - correct}")
print(f"accuracy--> {round(correct/total, 3)*100}")
```
