```python
plt.figure(figsize=(10,12))
dfa['SalePrice'].plot.kde(xlim = (dfa['SalePrice'].min(), dfa['SalePrice'].max()))
plt.axvline(dfa['SalePrice'].median(), color='blue', label='median')
plt.axvline(dfa['SalePrice'].mean(), color='red', label='mean')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,12))
dfa['Year Built'].plot.kde(xlim = (dfa['Year Built'].min(), dfa['Year Built'].max()))
plt.axvline(dfa['Year Built'].median(), color='blue', label='median')
plt.axvline(dfa['Year Built'].mean(), color='red', label='mean')
plt.axvline(2005, color='green', label='mode')
plt.legend()
plt.show()
```
