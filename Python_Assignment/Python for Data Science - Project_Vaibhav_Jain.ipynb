# Creating a db and table

```python
con = sqlite3.connect("data.db")
```

```python
cur = con.cursor()
```

```python
query = 'CREATE TABLE employee(ID integer, JobTitle text, EmailAddress VARCHAR(320), FirstNameLastName text, vaccinated text )'
```

```python
cur.execute(query)
```

# Creating a nested list from the json data

```python
for i in range(len(data_y)):
    list_temp = []
    for k,v in data_y[i].items():
        list_temp.append(v)
    list_data_y.append(list_temp)
```

```python
list_data_y
```

# Instering the Json data into the db

```python
query = 'INSERT into employee (ID, JobTitle, EmailAddress, FirstNameLastName, vaccinated) VALUES (?, ?, ?, ?, ?)'
for i in range(len(list_data_y)):
    cur.execute(query,list_data_y[i])
```

# Updating Bonus

```python
def bonus_percent(row):
    if row['vaccinated'] == 'True':
        bonus = 15
    elif row['vaccinated'] == 'False':
        bonus = 5
    return bonus
```

```python
dfa['bonus percecnt'] = dfa.apply(bonus_percent, axis=1)
```
